```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('amenity_data.csv')
print(data)
rqd_data = data[['name','amenity', 'latitude', 'longitude']]; #filtering amenities to get data that we want
rqd_data['latitude'] = rqd_data['latitude'].apply(lambda x: float(x));
rqd_data['longitude'] = rqd_data['longitude'].apply(lambda x: float(x));
rqd_data['amenityType'] = 'NULL';
```

```python
food = pd.array(['restaurant', 'pub','fast_food', 'bar', 'ice_cream', 'cafe', 'coffee_stand', 'food_court', 'bbq',
                             'concession_stand', 'brewery', 'food_sharing'])
place_of_worship = pd.array(['place_of_worship'])
post_box = pd.array(['post_box', 'post_office', 'letter_box', 'mailroom', 'Mailroom'])
car_services = pd.array(['fuel', 'car_wash', 'car_sharing', 'charging_station',
                         'car_rental'])
parking = pd.array(['parking_entrance', 'parking', 'parking_space', 'bicycle_parking', 'motorcycle_parking'])
entertainment = pd.array(['cinema', 'theatre','arts_centre','casino', 'hookah_lounge', 'nightclub','art_gallery', 'studio',
                          'event_space', 'events_venue', 'stripclub', 'art_classes', 'art_studio', 'yacht_club', 'lounge',
                          'makerspace', 'hall', 'swimming_pool', 'psychic', 'event_hall', 'amphitheatre', 'conference_centre',
                          'karaoke_box', 'lounger'])
finance = pd.array(['atm', 'bank', 'financial_advice', 'money_lender', 'money_transfer', 'investments'])
education = pd.array(['school', 'kindergarten', 'university', 'library', 'driving_school', 'college', 'prep_school', 'classes',
                      'trade_school', 'music_school', 'lessons', 'language_school', 'study_lounge', 'cosmetology_school',
                      'tutoring', 'tutor', 'nursing_school', 'public_bookcase'])
animals = pd.array(['veterinary', 'animal_boarding', 'animal_shelter', 'dog_grooming', 'animal_service', 'animal_training'])
rentals = pd.array(['boat_rental', 'ski_rental', 'bicycle_rental'])
playgrounds = pd.array(['picnic_area'])
negative_amenities = pd.array(['grave_yard', 'waste_disposal', 'funeral_home', 'sanitary_dump_station', 'prison', 'mortuary',
                               'waste_transfer_station', 'ranger_station', 'crematorium'])
storage = pd.array(['lockers', 'locker', 'food_storage', 'delivery_locker', 'boat_storage'])
self_care = pd.array(['community_centre', 'dojo', 'tanning', 'nail salon', 'spa', 'massage', 'health_club', 'medispa',
                      'training', 'public_bath', 'shampooer'])
shopping = pd.array(['vending_machine', 'marketplace'])
transportation = pd.array(['ferry_terminal', 'bus_station', 'taxi'])
emergency = pd.array(['police', 'fire_station', 'mountain_rescue'])
healthcare = pd.array(['pharmacy', 'doctors', 'fountain', 'hospital', 'dentist', 'clinic', 'orthodontics',
                       'alternative_medicine', 'chiropractic', 'nursing_home', 'doctors;dentist'])
child_services = pd.array(['childcare'])
not_needed = pd.array(['toilets', 'weighbridge', 'telephone', 'drinking_water','social_facility', 'social_centre', 
                        'shelter', 'bench', 'courthouse', 'waste_basket', 'vacuum', 'loading_dock',
                        'recycling', 'townhall', 'clock', 'hunting_stand', 'boat_sales', 'compressed_air',
                        'biergarten', 'apartments', 'binoculars', 'shower', 'table', 
                        'bicycle_repair_station', 'device_charging_station', 'vacuum_cleaner', 'payment_terminal',
                        'bureau_de_change', 'horse_parking', 'shoeshine', 'coworking_space', 'coworking_space',
                        'smoking_station', 'polling_station' 'watering_place', 'book_return', 'hitching_post', 'smoking_area',
                        'water_point', 'give_box', 'tool_library', 'chair'])
```

```python
def get_amenities(x):
    if(x in food):
        return 'food'
    elif(x in place_of_worship):
        return 'place_of_worship'
    elif(x in post_box):
        return 'post_box'
    elif(x in car_services):
        return 'car_services'
    elif(x in parking):
        return 'parking'
    elif(x in entertainment):
        return 'entertainment'
    elif(x in finance):
        return 'finance'
    elif(x in education):
        return 'education'
    elif(x in animals):
        return 'animals'
    elif(x in playgrounds):
        return 'playgrounds'
    elif(x in negative_amenities):
        return 'negative_amenities'
    elif(x in storage):
        return 'storage'
    elif(x in self_care):
        return 'self_care'
    elif(x in shopping):
        return 'shopping'
    elif(x in transportation):
        return 'transportation'
    elif(x in emergency):
        return 'emergency'
    elif(x in healthcare):
        return 'healthcare'
    elif(x in child_services):
        return 'child_services'
    else:
        return 'not_needed'
```

```python
rqd_data['amenityType'] = rqd_data.apply(lambda x: get_amenities(x.amenity), axis = 1);
rqd_data = rqd_data[rqd_data.amenityType != 'not_needed'];
rqd_data.to_csv('rqd_data.csv')
```
